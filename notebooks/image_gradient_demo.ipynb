```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter, gaussian_filter1d, sobel
```

```python
image_path = '/home/brysongray/Downloads/noisy_image.png'
I = plt.imread(image_path)
print(np.max(I), np.min(I))

plt.figure(figsize=(8,8))
plt.imshow(I)
```

```python
I_x = gaussian_filter1d(I, 3, axis=1, order=1)
I_x = (I_x - np.min(I_x))/(np.max(I_x)-np.min(I_x))
I_y = gaussian_filter1d(I, 3, axis=0, order=1)
I_y = (I_y - np.min(I_y))/(np.max(I_y)-np.min(I_y))
fig, axs = plt.subplots(2, figsize=(8,16))
axs[0].imshow(I_x)
axs[1].imshow(I_y)
```

```python
I_x = gaussian_filter(sobel(I, axis=1), 3)
I_x = (I_x - np.min(I_x))/(np.max(I_x)-np.min(I_x))
I_y = gaussian_filter(sobel(I, axis=0), 3)
I_y = (I_y - np.min(I_y))/(np.max(I_y)-np.min(I_y))
fig, axs = plt.subplots(2, figsize=(8,16))
axs[0].imshow(I_x)
axs[1].imshow(I_y)
```

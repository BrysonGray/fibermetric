# Circular Harmonics

We want to fit a fiber orientation distribution function representing the density of myelinated fibers in a given direction to image gradients on myelin stained histology. We will use fourier series to represent the distribution. 

$$s_N(x) = \frac{a_0}{2} + \sum_{n=1}^{N}\left(a_ncos\left(\frac{2\pi}{P}nx\right) + b_nsin\left(\frac{2\pi}{P}nx\right)\right)$$ in this case the period, P, is $2\pi$

Given a distribution of gradients we will minimize 

```python
import numpy as np
import matplotlib.pyplot as plt

N = 25

f = np.random.normal(np.pi/2, 0.5, 30)

fig, ax = plt.subplots()

plt.plot(f)
```

```python
plt.hist(f,bins=30)
```

```python
F = np.fft.rfft(f, n=100)
```

```python
len(F)
```

```python
plt.plot(F)
```

```python
f_ = np.fft.irfft(F,n=100)
len(f_)
```
